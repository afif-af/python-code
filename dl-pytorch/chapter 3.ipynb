```python
img_gray_weighted_fancy=torch.einsum('...chw,c->...hw',img_t,weights)
batch_gray_weighted_fancy=torch.einsum('...chw,c->...hw',batch_t,weights)
batch_gray_weighted_fancy.shape
```
